```python
sns.set_style("whitegrid")
ERROR_LABELS = {"Cifar100 Error": "Cifar100", "Cifar100-C Error": "Cifar100-C"}

for network in df_std["Network"].unique():
    fig, axes = plt.subplots(1, 2, figsize=(15.5, 5.5))
    #df_part = df_prune.query(f"Network == '{network}' & not Amda")

    for (error, label), ax in zip(ERROR_LABELS.items(), axes.flatten()):
        sns.lineplot(data=df_std, x="Model Size", y=error , hue="Prune Method", ax=ax, marker="s", style="Prune Method")

        #y, x = df_baseline.query(f"Network == '{network}' & not Amda")[[error, "Model Size"]].to_numpy()[0]

        ax.scatter(df_std_baseline["Model Size"], df_std_baseline[error], marker='x', color='black', label='Original Model', )
        plt.tight_layout()

        handles, labels = ax.get_legend_handles_labels()
        ax.legend('', frameon=False)
        ax.set_title(label)
        ax.set_ylabel(error + " (%)" )
        ax.set_xlabel("Num. of Parameters (millions)")

    fig.tight_layout(pad=2.5)
    fig.legend(handles, labels, loc="lower center", borderaxespad=0., ncol=5)
    plt.savefig(f"figures/pruning_cifar_plot_std.pdf")
```

```python
df_aug_no_kd=df_aug.query("`Prune Method`=='Global Weight' or `Prune Method`=='L¹ Filter'")
df_kd_aug_plot=df_kd_aug.append(df_aug_no_kd)
```

```python
df_kd_aug_plot.iloc[[-1, -2],[8]]="KnowledgeDistill"
#df_kd_aug_plot.at(-1, "Loss")="KnowledgeDistill"
#df_kd_aug_plot.at(-2, "Loss")="KnowledgeDistill"
df_kd_aug_plot=df_kd_aug_plot.replace("KnowledgeDistill", "KD")
df_kd_aug_plot=df_kd_aug_plot.replace("StandardLoss", "No Distill.")
df_kd_aug_plot["Distillation Method"]=df_kd_aug_plot["Loss"]
```

```python
sns.set_style("whitegrid")
ERROR_LABELS = {"Cifar100 Error": "Cifar100", "Cifar100-C Error": "Cifar100-C"}

for network in df_kd_aug_plot["Network"].unique():
    fig, axes = plt.subplots(1, 2, figsize=(15.5, 5.5))
    #df_part = df_prune.query(f"Network == '{network}' & not Amda")

    for (error, label), ax in zip(ERROR_LABELS.items(), axes.flatten()):
        sns.lineplot(data=df_kd_aug_plot, x="Model Size", y=error , hue="Prune Method", ax=ax, marker="s", style="Distillation Method")

        #y, x = df_baseline.query(f"Network == '{network}' & not Amda")[[error, "Model Size"]].to_numpy()[0]

        ax.scatter(df_aug_baseline["Model Size"], df_aug_baseline[error], marker='x', color='black', label='Original Model', )
        plt.tight_layout()

        handles, labels = ax.get_legend_handles_labels()
        ax.legend('', frameon=False)
        ax.set_title(label)
        ax.set_ylabel(error + " (%)" )
        ax.set_xlabel("Num. of Parameters (millions)")

    fig.tight_layout()
    fig.legend(handles[:-1], labels[:-1], bbox_to_anchor=(0.36, 1.2), loc=2, borderaxespad=0., ncol=2)
    plt.savefig(f"figures/pruning_cifar_plot_kd_aug.pdf", bbox_inches='tight')
```
