```python
from sklearn.manifold import TSNE
from sklearn.decomposition import PCA
from scipy.special import softmax

from matplotlib.pyplot import cm

from random import sample
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_pickle("dataframes/short_features_logits.pkl")
```

```python
df["Softmax Distribution"] = [softmax(l / 10) for l in df["Logits"]]
```

```python
def corr_plot(softmax_distr, **kwargs):
    correlations = pd.DataFrame(np.stack(softmax_distr)).corr()
    sns.heatmap(correlations, vmin=-1, vmax=1, **kwargs)
```

```python
df["Net"] = df["Network Name"]
df["Distr."] = df["Distribution"]
```

```python
g = sns.FacetGrid(df, col="Net", row="Distr.")
g.map(corr_plot, "Softmax Distribution", cmap=cm.seismic, cbar=False)
```

```python
g = sns.FacetGrid(df, col="Net", row="Distr.")
g.map(corr_plot, "Softmax Distribution", cmap=cm.seismic, cbar=False)
```
