```python
from pruneshift.datamodules import datamodule
from pruneshift.networks import network
from pruneshift.modules import VisionModule
import pytorch_lightning as pl
import os
import logging

DATA_PATH = os.environ["DATASET_PATH"]
MODEL_PATH = os.environ["MODEL_PATH"]
logging.basicConfig(level=logging.INFO)
```

```python
dm = datamodule("cifar10_corrupted", DATA_PATH)
dm_orig = datamodule("cifar10", DATA_PATH)
```

```python
trainer = pl.Trainer(gpus=1)
```

```python
net = VisionModule(network("cifar10_resnet18", model_path=MODEL_PATH), test_labels=dm.labels)
```

```python
trainer.test(net, datamodule=dm)
```

```python
net = VisionModule(network("cifar10_resnet18", model_path=MODEL_PATH))
trainer.test(net, datamodule=dm_orig)
```
