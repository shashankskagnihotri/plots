```python
df_comb_special = pd.concat([df_baseline.query("Amda"), df_special_plot])
```

```python
df_comb_special["Network"] = df_comb_special["Network"].map({"mobilenet_v2": "MobileNetV2", "resnet18": "ResNet18", "mnasnet1_0": "MnasNet"})
```

```python
fig, axes = plt.subplots(1, 3, figsize=(17.5, 5))

for ax, error in zip(axes.flatten(), ["ImageNet100 Error", "ImageNet100-C Error", "ImageNet100-R Error"]):
    sns.lineplot(data=df_comb_special, x="Model Size", y=error, style="Network", hue="Network", marker="o", ax=ax)
    # ax.legend(loc="lower center", borderaxespad=-6., ncol=3, )
    ax.set_xlabel("Num. of Parameters (millions)")
    ax.set_ylabel(error + " (%)")
    ax.set_xlim(0, 3.5)
    ax.set_title(error.split(" ")[0])
    # ax.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
    if error != "ImageNet100-R Error":
        ax.legend('', frameon=False)
    pts = []
    for network in ["mnasnet1_0", "mobilenet_v2"]:
        y, x = df_baseline.query(f"Amda & Network == '{network}'")[[error, "Model Size"]].to_numpy()[0]
        pts.append(np.array([x, y]))
        # ax.scatter(x, y, zorder=19)
    
    mid = (pts[0] + pts[1]) / 2 + np.array([-1, 10])

    for coords in pts:
        ax.annotate('Unpruned Models', xy=coords, xytext=mid, color="grey",
                    arrowprops=dict(facecolor='grey', shrink=0.05, alpha=0.7),
                    )
plt.tight_layout()
plt.savefig(f"figures/hydra_model_size_tradeoff.pdf")
```
