```python
# We try the most simple approach to prune and see how it works with CIFAR-C.
from pruneshift.datasets import CIFAR100C, CIFAR10C, ImageNetC
from pruneshift.datamodules import ShiftDataModule as datamodule
import os
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import torch

label=0
DATA_PATH = os.environ["DATASET_PATH"]
IMAGENET_PATH = "/work/dlclarge2/agnihotr-shashank-pruneshift/hoffmaja-pruneshift/datasets/ILSVRC2012-100/train"
IMAGENET_CORRUPTED = "/work/dlclarge2/agnihotr-shashank-pruneshift/hoffmaja-pruneshift/datasets/ILSVRC2012-100/"
DEEPAUGMENT= "/data/datasets/DeepAugment/EDSR"

def augmix_vis(name, n, path=DATA_PATH):
    dm = datamodule(name, path, with_normalize=False)
    fig, axes = plt.subplots(n, 3, figsize=(8, 8))
    dm.prepare_data()
    dm.setup("fit")
    data_iter = iter(dm.train_dataloader())
    for lvl, row in enumerate(axes):
        imgs, y = next(data_iter)
        for idx, ax in enumerate(row):
            img = (255 * imgs[idx][0]).permute([1, 2, 0]).to(torch.uint8).numpy()
            ax.imshow(Image.fromarray(img))
            ax.axis('off')

def vis(name, n, path=DATA_PATH):
    dm = datamodule(name, path, with_normalize=False)
    fig, axes = plt.subplots(n, 3, figsize=(8, 8))
    dm.prepare_data()
    dm.setup("fit")
    data_iter = iter(dm.train_dataloader())
    for lvl, row in enumerate(axes):
        for idx, ax in enumerate(row):
            imgs, y = next(data_iter)
            img = (255 * imgs[0]).permute([1, 2, 0]).to(torch.uint8).numpy()
            ax.imshow(Image.fromarray(img))
            ax.axis('off')

def vis_labels(name, path=DATA_PATH):
    dm = datamodule(name, path, batch_size=1000)
    dm.setup("fit")
    _, y_batch = next(iter(dm.train_dataloader()))
    plt.hist(np.array(y_batch))
    

def corruption(dataset_cls, corruptions=["snow", "brightness", "pixelate", "glass_blur", "fog", "jpeg_compression"], path=DATA_PATH):
    images = []
    for corr in corruptions:
        data = dataset_cls(path, corr).lvl_subsets()[-1]
        images.append(data[-320][0])
        label=data[0][1]
    
    fig, axes = plt.subplots(int(len(images)/6), 6, figsize=(12, 20))
    it = range(len(images))
    axes.flatten()
    print(axes)
    i=0
    for a in axes:
        #for a in ax:
        a.imshow(images[i])
        a.axis('off')
        a.set_title(corruptions[i])
        i+=1
    fig.tight_layout()
        
        

    #for label, ax, img, i in zip(corruptions, axes, images,it):
     #   print(ax)
      #  ax.imshow(img)
       # ax.axis('off')
        #ax.set_title(label)
```
